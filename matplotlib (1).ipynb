```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. 
# Using this generated data plot a scatter plot.
```

```python
# ans: Scatter plots are used to observe relationship between variables and uses dots 
# o represent the relationship between them. The scatter() method in the matplotlib 
# ibrary is used to draw a scatter plot. Scatter plots are widely 
# used to represent relation among variables and how change in one affects the other. 
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("This is x-axis")
plt.ylabel("This is y-axis")
plt.title("THIS IS SCATTER GRAPH")
plt.scatter(x,y)
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
```

```python
#ans:: subplot() function adds subplot to a current figure at the specified grid position.  It is similar to the subplots() function however unlike subplots() it adds one subplot at a time. So to create multiple plots you will need several lines of code with the subplot() function.
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(121)
plt.plot(x, y, color="orange", marker="*")
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(121)
plt.plot(x, y, color="yellow", marker="*")
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(121)
plt.plot(x, y, color="pink", marker="*")
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(121)
plt.plot(x, y, color="purple", marker="*")
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
```

```python
#A bar plot is a graph that shows the relationship between a numeric and 
#a categorical variable using rectangular bars. The length or height of each bar
#is proportional to the value it represents. A bar plot can be vertical or horizontal


company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.xlabel("company")
plt.ylabel("profit")
plt.bar(company,profit,color="pink")
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
```

```python
# Box Plot: It is a type of chart that depicts a group of numerical data through 
# their quartiles. It is a simple way to visualize the shape of our data. 
# It makes comparing characteristics of data between categories very easy.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot(box1)
plt.boxplot(box2)

```

```python
# Que 1: Name anyfive plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
# 1: stripplot

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
iris = sns.load_dataset('iris')

ax = sns.stripplot(x='species', y='sepal_length', data=iris)
  
plt.title('Graph')
plt.show()
```

```python
#2: Swarmplot 

ax = sns.swarmplot(x='species', y='sepal_length', data=iris)
plt.title('Graph')
plt.show()
```

```python
#3:Barplot
df = sns.load_dataset('tips')
sns.barplot(x ='sex', y ='total_bill', data = df,  
            palette ='plasma')
```

```python
#4: countplot
df = sns.load_dataset('tips') 
  
sns.countplot(x ='sex', data = df) 
```

```python
# Boxplot
# Box Plot is the visual representation of the depicting groups of numerical data through their quartiles. Boxplot is also used to detect the outlier in the data set.

```

```python
df = sns.load_dataset('tips')
sns.boxplot(x='day', y='total_bill', data=df, hue='smoker')
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(data=fmri, kind="line", x="timepoint", y="signal")
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
# 'pclass', y = 'age' and y = 'fare'.
```

```python
titanic=sns.load_dataset("titanic")
sns.barplot(data=titanic,x="pclass",y="fare")
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.
```

```python
df=sns.load_dataset("diamonds")
sns.histplot(data=df,y='price',hue='cut')
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.
```

```python
df=sns.load_dataset('iris')
sns.pairplot(data=df,hue='species')
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
```
